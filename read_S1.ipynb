```python
import os
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define the base directory (current working directory)
base_dir = os.getcwd()  #used to check where you are right now

print("base_dir:", base_dir)

working_path = '/run/user/1000/gvfs/smb-share:server=ifu-eo-cds-1.ethz.ch,share=working/Marion/AGRI4SAR/Sentinel-1/Pre_processed/subset_0_of_S1A_IW_SLC__1SDV_20250215T172341_20250215T172408_057910_072512_E067_Orb_split_tnr_Cal_deb_TC.data/' #FOLDER of your data
vv_img_path = working_path + "Sigma0_IW1_VV.img"
vh_img_path = working_path + "Sigma0_IW1_VH.img"
```

```python
with rasterio.open(vv_img_path) as vv_dataset, rasterio.open(vh_img_path) as vh_dataset:
    vv_band = vv_dataset.read(1)  
    vh_band = vh_dataset.read(1)  
    
    print("VV Image Shape:", vv_band.shape)
    print("VH Image Shape:", vh_band.shape)

    # Display the image using matplotlib
    plt.imshow(vv_band, cmap='gray', vmin=0, vmax=0.9)  # Use a suitable colormap
    plt.colorbar()  # Add color bar
    plt.title("Raster Image")
    plt.show()
```

```python
# plot the vv and vh channels
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# Plot VV band
ax[0].imshow(vv_band, cmap="gray", vmin=0, vmax=0.9)
ax[0].set_title("VV Band")
ax[0].axis("off")

# Plot VH band
ax[1].imshow(vh_band, cmap="gray", vmin=0, vmax=0.9)
ax[1].set_title("VH Band")
ax[1].axis("off")
```
